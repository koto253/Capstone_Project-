Column Descriptions for Economic Dataset
1. Country

    Description: The name of the country.
    Details: This column contains the names of the countries included in the dataset.

2. Year

    Description: The year of the data.
    Details: This column indicates the specific year for which the data was recorded.

3. Exports_USD

    Description: Exports of goods and services (current US$).
    Details: The total value of all goods and services exported by a country, measured in current U.S. dollars.

4. GDP_USD

    Description: Gross Domestic Product (GDP) in current US$.
    Details: The total market value of all final goods and services produced within a country in a given year, measured in current U.S. dollars.

5. GDP_Capita_USD

    Description: GDP per capita in current US$.
    Details: The GDP of a country divided by its population, giving the average economic output per person, measured in current U.S. dollars.

6. Capital_Formation_USD

    Description: Gross capital formation (current US$).
    Details: The total value of a country's investments in fixed assets and net changes in inventories, measured in current U.S. dollars.

7. Imports_USD

    Description: Imports of goods and services (current US$).
    Details: The total value of all goods and services imported by a country, measured in current U.S. dollars.

8. Industry_Value_Added_USD

    Description: Industry (including construction), value added (current US$).
    Details: The net output of a country’s industrial sector (including construction) after adding up all outputs and subtracting intermediate inputs, measured in current U.S. dollars.

9. Labor_Force

    Description: Labor force, total.
    Details: The total number of people who are of working age and are either employed or actively seeking employment.

10. Adult_Lit_Rate

    Description: Literacy rate, adult total (% of people ages 15 and above).
    Details: The percentage of people aged 15 and above who can read and write.

11. Military_Exp_USD

    Description: Military expenditure (current USD).
    Details: The total amount of money spent by a country on its military forces, measured in current U.S. dollars.

12. Pop_Growth_Percent

    Description: Population growth (annual %).
    Details: The annual percentage increase in a country’s population.

13. Population_Total

    Description: Population, total.
    Details: The total number of people living in a country.

14. Research_GDP%

    Description: Research and development expenditure (% of GDP).
    Details: The total expenditure on research and development activities, expressed as a percentage of the country’s GDP.

15. Trade_Percent_GDP

    Description: Trade (% of GDP).
    Details: The sum of exports and imports of goods and services, measured as a share of gross domestic product.

16. Unemployment_Percent

    Description: Unemployment, total (% of total labor force) (national estimate).
    Details: The percentage of the total labor force that is unemployed and actively seeking employment.
